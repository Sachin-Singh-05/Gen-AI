## Currency Conversion Tool and Tool Calling

```python
# from langchain_core.tools import InjectedToolArg
from typing import Annotated
from langchain_core.tools import tool
from langchain_core.tools.base import InjectedToolArg


@tool
def get_conversion_factor(base_currency: str, target_currency: str) -> float:
  """
  This tool brings an exchange rate from base currency to target currency via an api.
  """

  url = f"https://v6.exchangerate-api.com/v6/c22231851d338cd7da4cb8e3/pair/{base_currency}/{target_currency}"

  response = requests.get(url)

  return response.json()


@tool
def convert(base_currency_value : int , conversion_rate : Annotated[float,InjectedToolArg]) -> float :
  """
  This tool converts base currency value to target currency value.
  """

  return base_currency_value * conversion_rate
```

```python
convert.args
```

```python
get_conversion_factor.invoke({'base_currency': 'USD', 'target_currency': 'INR'})
```

```python
rate = get_conversion_factor.invoke({'base_currency': 'USD', 'target_currency': 'INR'})
rate['conversion_rate']
```

```python

convert.invoke({'base_currency_value': 10, 'conversion_rate': rate['conversion_rate']})
```

```python
print(type(rate['conversion_rate']))
```

```python
# tool binding
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)
```

```python
llm_with_tools_1 = llm.bind_tools([get_conversion_factor,convert])
```

```python
message = [HumanMessage('what is the conversion factor between USD and INR , then convert 10 USD to INR')]
```

```python
message
```

```python
ai_message = llm_with_tools_1.invoke(message)
```

```python
ai_message
```

```python
message.append(ai_message)
```

```python
ai_message.tool_calls
```

```python
import json

for tool_call in ai_message.tool_calls:
  if tool_call['name'] == 'get_conversion_factor':
    tool_message_1 = get_conversion_factor.invoke(tool_call)
    conversion_rate = json.loads(tool_message_1.content)['conversion_rate']
    message.append(tool_message_1)
  if tool_call['name'] == 'convert':
    tool_call['args']['conversion_rate'] = conversion_rate
    tool_message_2 = convert.invoke(tool_call)
    message.append(tool_message_2)


```
